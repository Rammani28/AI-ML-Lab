Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('restaurant_data.csv')
df.head(5)
```

```python
df.describe()
```

```python
# Parking Availability

df.groupby('Parking Availability').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
```

```python
sns.countplot(data=df, x='Location')
df['Location'].value_counts()
```

```python
sns.countplot(data=df, x="Parking Availability")
df['Parking Availability'].value_counts()
```

```python
# pie chart
df['Location'].value_counts().plot(kind='pie',autopct='%0.1f')

```
